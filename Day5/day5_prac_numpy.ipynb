Practice numpy operations

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales_data = np.array([
    [120, 150, 100, 90],   # Product 1
    [200, 180, 160, 170],  # Product 2
    [90,  110, 130, 120],  # Product 3
    [300, 280, 250, 270],  # Product 4
    [150, 170, 140, 160]   # Product 5
])
print("--Sales Analysis")
print("data shape:", sales_data.shape)
print("first 3 products: \n ", sales_data[:3] )
```

```python
#total sales
print(np.sum(sales_data, axis=0))
y_total = np.sum(sales_data[:,0:], axis=0)
print(y_total)
```

```python
# mimimum sales per product
min_sp =np.min(sales_data, axis=1)
print(min_sp)
# maximum sales per prooduct
max_sp =np.max(sales_data, axis=1)
print(max_sp)
# Minimum sales per day
min_spd = np.min(sales_data, axis=0)
print(min_spd)

#Average sale per product
avg_sale =np.mean(sales_data[:3], axis=1)
print(avg_sale)
```

```python

# Comulative sum
comsum = np.cumsum(sales_data, axis=1)
print(comsum)
# Graph plot
plt.figure(figsize=(4,4))
plt.plot(np.mean(comsum, axis=0))
plt.title("sale data per day")
plt.xlabel("days")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1,2,3,4,5,6,7,8])
vector2 = np.array([3,4,5,6,7,8,9,1])

print("Vector Addition: ", vector1 + vector2)
print("\n Multiplication:", vector1 * vector2)

print("Dot product:",np.dot(vector1, vector2))

angel = np.arccos(np.dot(vector1, vector2) / ((np.linalg.norm(vector1))*
                                              (np.linalg.norm(vector2))))
print("Angel",angel)
```

```python
# Vectorized Operations
R_type = np.array(['Pizza', 'Chinese','Burger','Cafe'])
v_arr = np.vectorize(str.upper)
print("Vectorized array",v_arr(R_type))

# Broadcasting
day_sale = sales_data / 30
print(sales_data)
print("Per day sales \n ",day_sale)
```
